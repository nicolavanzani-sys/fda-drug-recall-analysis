```python
import requests 
import pandas as pd
```

```python
# Importazione dati
def importazione(link, limit):
    data_list=[]
    skip = 0
    for x in range(0, 18):
        data = requests.get(f'{link}?skip={skip}&limit={limit}')
        data_list+=data.json()['results']
        skip+=limit
    
    return data_list 

dati = importazione(link = 'https://api.fda.gov/drug/enforcement.json', limit=1000)

print(len(dati))
```

```python
# Visualizzazione DataFrame
df = pd.DataFrame(dati)
print(df.shape)
print(df.columns.tolist())
type(df)
```

```python
# Esportazione dei risultati
df.to_csv('fda_data.csv')
```
