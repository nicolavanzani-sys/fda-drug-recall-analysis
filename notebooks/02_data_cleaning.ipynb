```python
import pandas as pd
import datetime as dt
import numpy as np
import datetime as dt
import ast
```

```python
# Importo il dataset
df=pd.read_csv('/Users/nicolavanzani/fda-drug-recall-analysis/data/01_fda_data', index_col= False)
```

```python
# Controllo le tipologie di dati nelle colonne
print(df.dtypes)
```

```python
# Verifico la presenzadi valori nulli
print(df.isnull().sum())
```

```python
# Eliminazione di colonne superflue
df = df.drop(columns=['more_code_info', 'address_1', 'address_2', 'postal_code', 'Unnamed: 0'])
```

```python
# Conversione in date
date_cols = [col for col in df.columns if col.endswith('date')]
for col in date_cols:
    df[col] = pd.to_datetime(
        df[col].fillna(0).astype('Int64').astype(str),
        format='%Y%m%d',
        errors='coerce'
        )
```

```python
# Visualizzazione dei dati in openfda
print(df['openfda'].value_counts().head(10))
```

```python
# Estrazione dei dati da openfda
def estrai_brand(x):
    dizionario = ast.literal_eval(x)
    if 'brand_name' in dizionario:
        return dizionario['brand_name'][0]  # primo elemento della lista
    else:
        return None

df['brand_name'] = df['openfda'].apply(estrai_brand)

print(df['brand_name'].value_counts().head(5))
print(df['brand_name'].isnull().sum())
```

```python
# Eliminazione delle colonne in eccesso
df = df.drop(columns = ['openfda'])
```
